```python
def encode_chromosome(course_name, lecture_type, section, section_strength, professor, first_lecture_day, first_lecture_timeslot, first_lecture_room, first_lecture_room_size, second_lecture_day, second_lecture_timeslot, second_lecture_room, second_lecture_room_size):
    binary_encoded_chromosome = []

    course_encoding = [1 if course == course_name else 0 for course in Courses]
    binary_encoded_chromosome.extend(course_encoding)

    lecture_type_encoding = [1 if type_ == lecture_type else 0 for type_ in ["theory", "lab"]]
    binary_encoded_chromosome.extend(lecture_type_encoding)

    section_encoding = [1 if section_ == section else 0 for section_ in Sections]
    binary_encoded_chromosome.extend(section_encoding)

    section_strenth_binary = [int(bit) for bit in bin(section_strength)[2:].zfill(7)]
    binary_encoded_chromosome.extend(section_strenth_binary)

    professor_encoding = [1 if prof == professor else 0 for prof in Professors]
    binary_encoded_chromosome.extend(professor_encoding)

    first_lecture_day_encoding = [1 if day == first_lecture_day else 0 for day in ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]]
    binary_encoded_chromosome.extend(first_lecture_day_encoding)

    first_lecture_timeslot_encoding = [1 if timeslot == first_lecture_timeslot else 0 for timeslot in Timeslots]
    binary_encoded_chromosome.extend(first_lecture_timeslot_encoding)

    first_lecture_room_encoding = [1 if room == first_lecture_room else 0 for room in Rooms]
    binary_encoded_chromosome.extend(first_lecture_room_encoding)

    first_lecture_room_size_encoding = [1 if room_size == first_lecture_room_size else 0 for room_size in [60, 120]]
    binary_encoded_chromosome.extend(first_lecture_room_size_encoding)

    second_lecture_day_encoding = [1 if day == second_lecture_day else 0 for day in ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]]
    binary_encoded_chromosome.extend(second_lecture_day_encoding)

    second_lecture_timeslot_encoding = [1 if timeslot == second_lecture_timeslot else 0 for timeslot in Timeslots]
    binary_encoded_chromosome.extend(second_lecture_timeslot_encoding)

    second_lecture_room_encoding = [1 if room == second_lecture_room else 0 for room in Rooms]
    binary_encoded_chromosome.extend(second_lecture_room_encoding)

    second_lecture_room_size_encoding = [1 if room_size == second_lecture_room_size else 0 for room_size in [60, 120]]
    binary_encoded_chromosome.extend(second_lecture_room_size_encoding)

    return binary_encoded_chromosome
```

```python
def decode_course(chromosome):
    index = 0
    course_index = chromosome[index:index + len(Courses)]
    return Courses[course_index.index(1)]

def decode_course_type(chromosome):
    index = len(Courses)
    type_index = chromosome[index:index + 2]
    return 'Theory' if type_index[0] == 1 else 'Lab'

def decode_section(chromosome):
    index = len(Courses) + 2
    section_index = chromosome[index:index + len(Sections)]
    return Sections[section_index.index(1)]

def decode_section_strength(chromosome):
    index = len(Courses) + 2 + len(Sections)
    section_strength_index = chromosome[index:index + 7]
    section_strength = int(''.join(map(str, section_strength_index)), 2)
    return section_strength

def decode_professor(chromosome):
    index = len(Courses) + 2 + len(Sections) + 7
    professor_index =  chromosome[index:index + len(Professors)]
    return Professors[professor_index.index(1)]

def decode_first_lecture_day(chromosome):
    index = len(Courses) + 2 + len(Sections) + 7 + len(Professors)
    first_lecture_day_index = chromosome[index:index + 5]
    return ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"][first_lecture_day_index.index(1)]

def decode_first_lecture_timeslot(chromosome):
    index = len(Courses) + 2 + len(Sections) + 7 + len(Professors) + 5
    first_leture_timeslot_index = chromosome[index:index + len(Timeslots)]
    return Timeslots[first_leture_timeslot_index.index(1)]

def decode_first_lecture_room(chromosome):
    index = len(Courses) + 2 + len(Sections) + 7 + len(Professors) + 5 + len(Timeslots)
    first_lecture_room_index = chromosome[index:index + len(Rooms)]
    return Rooms[first_lecture_room_index.index(1)]

def decode_first_lecture_room_size(chromosome):
    index = len(Courses) + 2 + len(Sections) + 7 + len(Professors) + 5 + len(Timeslots) + len(Rooms)
    first_lecture_room_size_index = chromosome[index:index + 2]
    return 60 if first_lecture_room_size_index[0] == 1 else 120

def decode_second_lecture_day(chromosome):
    index = len(Courses) + 2 + len(Sections) + 7 + len(Professors) + 5 + len(Timeslots) + len(Rooms) + 2
    second_lecture_day_index = chromosome[index:index + 5]
    return ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"][second_lecture_day_index.index(1)]

def decode_second_lecture_timeslot(chromosome):
    index = len(Courses) + 2 + len(Sections) + 7 + len(Professors) + 5 + len(Timeslots) + len(Rooms) + 2 + 5
    second_lecture_timeslot_index = chromosome[index:index + len(Timeslots)]
    return Timeslots[second_lecture_timeslot_index.index(1)]

def decode_second_lecture_room(chromosome):
    index = len(Courses) + 2 + len(Sections) + 7 + len(Professors) + 5 + len(Timeslots) + len(Rooms) + 2 + 5 + len(Timeslots)
    second_lecture_room_index = chromosome[index:index + len(Rooms)]
    return Rooms[second_lecture_room_index.index(1)]

def decode_second_lecture_room_size(chromosome):
    index = len(Courses) + 2 + len(Sections) + 7 + len(Professors) + 5 + len(Timeslots) + len(Rooms) + 2 + 5 + len(Timeslots) + len(Rooms)
    second_lecture_room_size_index = chromosome[index:index + 2]
    return 60 if second_lecture_room_size_index[0] == 1 else 120


def decode_chromosome(chromosome):
    decoded_chromosome = []
    decoded_chromosome.append(decode_course(chromosome))
    decoded_chromosome.append(decode_course_type(chromosome))
    decoded_chromosome.append(decode_section(chromosome))
    decoded_chromosome.append(decode_section_strength(chromosome))
    decoded_chromosome.append(decode_professor(chromosome))
    decoded_chromosome.append(decode_first_lecture_day(chromosome))
    decoded_chromosome.append(decode_first_lecture_timeslot(chromosome))
    decoded_chromosome.append(decode_first_lecture_room(chromosome))
    decoded_chromosome.append(decode_first_lecture_room_size(chromosome))
    decoded_chromosome.append(decode_second_lecture_day(chromosome))
    decoded_chromosome.append(decode_second_lecture_timeslot(chromosome))
    decoded_chromosome.append(decode_second_lecture_room(chromosome))
    decoded_chromosome.append(decode_second_lecture_room_size(chromosome))
    return decoded_chromosome
```

```python
#conflict functions
def same_room_during_same_timeslot(course_name, day, timeslot, room, room_usage, conflict_details):
    conflicts = 0
    if (day, timeslot, room) in room_usage:
        conflicts += 1
        conflict_details.append(f"{room} is already ocupied at {timeslot} on {day}")
    else:
        room_usage[(day, timeslot, room)] = course_name

    return conflicts

def sufficient_room_capacity(section_strength, capacity, room, conflict_details):
    conflicts = 0
    if section_strength > capacity:
        conflicts += 1
        conflict_details.append(f"{room} with capacity {capacity} is too small for {section_strength} students")

    return conflicts

def professor_course_count(professor, professor_courses, conflict_details):
    conflicts = 0
    if professor in professor_courses:
        professor_courses[professor] += 1
    else:
        professor_courses[professor] = 1
    
    if professor_courses[professor] > 3:
            conflicts += 1
            conflict_details.append(f"{professor} is assigned more than 3 courses.")

    return conflicts

def section_course_count(course_name, section, section_courses, conflict_details):
    conflicts = 0
    if section in section_courses:
        section_courses[section].add(course_name)
    else:
        section_courses[section] = {course_name}

    if len(section_courses[section]) > 5:
        conflicts += 1
        conflict_details.append(f"Section {section} has more than 5 courses in this semester. ")

    return conflicts

def professor_assigned_to_two_lectures_same_time(day, timeslot, population, professor, conflict_details):
    conflicts = 0
    unique_conflicts = set()

    for c in population:
        decoded_c = decode_chromosome(c)
        if professor == decoded_c[4] and ((timeslot == decoded_c[6] and day == decoded_c[5]) or (timeslot == decoded_c[10] and day == decoded_c[9])):
            unique_conflicts.add(f"{professor} is assigned two lectures at the same time at {timeslot} on {day}")  

    conflicts += len(unique_conflicts)
    
    for curr_conflict in unique_conflicts:
        conflict_details.append(curr_conflict)

    return conflicts

    
def same_section_assigned_different_rooms_at_same_time(day, timeslot, room, population, section, conflict_details):
    conflicts = 0
    unique_conflicts = set()

    for c in population:
        decoded_c = decode_chromosome(c)
        if section == decoded_c[2] and ((timeslot == decoded_c[6] and day == decoded_c[5] and room != decoded_c[7]) or 
                                    (timeslot == decoded_c[10] and day == decoded_c[9] and room != decoded_c[11])): 
            unique_conflicts.add(f"Section {section} is assigned different rooms at same time at {timeslot} on {day}")

    conflicts += len(unique_conflicts)

    for curr_conflict in unique_conflicts:
        conflict_details.append(curr_conflict)
        
    return conflicts


def same_room_assigned_to_different_sections_at_same_time(day, timeslot, room, population,  section, conflict_details):
    conflicts = 0
    unique_conflicts = set()

    for c in population:
        decoded_c = decode_chromosome(c)
        if room == decoded_c[7] and ((timeslot == decoded_c[6] and day == decoded_c[5] and room == decoded_c[7]) or
                (timeslot == decoded_c[10] and day == decoded_c[9] and room == decoded_c[11]) and section != decoded_c[2]):
            unique_conflicts.add(f"Section {section} is assigned to same rooms at the same time at {timeslot} on {day}")

    conflicts += len(unique_conflicts)

    for curr_conflict in unique_conflicts:
        conflict_details.append(curr_conflict)

    return conflicts


def lab_lectures_two_consective_time_slots(lecture_type, timeslot1, timeslot2, conflict_details):
    conflicts = 0
    if lecture_type.lower() == 'lab' and timeslot2 is not None:
        timeslot1_endhour, timeslot1_endminute = map(int, timeslot1.split('-')[1].split(':'))
        timeslot2_starthour, timeslot2_startminute = map(int, timeslot2.split('-')[0].split(':'))

        timeslot1_minutes = timeslot1_endhour * 60 + timeslot1_endminute
        timeslot2_minutes = timeslot2_starthour * 60 + timeslot2_startminute

        if (timeslot1_minutes + 15 != timeslot2_minutes):
            conflicts += 1
            conflict_details.append(f"Slot1 : {timeslot1} and Slot2 : {timeslot2} for lab are not consecutive")

    return conflicts        

def course_have_two_lectures_per_week_not_on_same_or_adjacent_days(course, population, conflict_details):
    conflicts = 0
    course_schedule = set()
    for c in population:
        decoded_c = decode_chromosome(c)
        if decoded_c[0] == course:
            course_schedule.add((decoded_c[5]))

    indexing = {"Monday": 1, "Tuesday": 2, "Wednesday": 3, "Thursday": 4, "Friday": 5}

    if len(course_schedule) == 2:
        schedule_days = [day for day in course_schedule] 

        if abs(indexing[schedule_days[0]]) - abs(indexing[schedule_days[1]]) > 1:
            conflicts += 1
            conflict_details.append(f"{course} have lectures on the same or adjacent days")
    else:
        conflicts += 1
        conflict_details.append(f"{course} have more than two scheduled lectures")

    return conflicts

    
```

```python
def calculate_conflicts(chromosome, population, room_usage, professor_courses, section_courses, conflict_details):
    conflicts = 0

    #Hard constraints Implementation
    
    # # 1 Classes can only be scheduled in free classroom.
    conflicts += same_room_during_same_timeslot(decode_course(chromosome), decode_first_lecture_day(chromosome), decode_first_lecture_timeslot(chromosome), decode_first_lecture_room(chromosome), room_usage, conflict_details)
    conflicts += same_room_during_same_timeslot(decode_course(chromosome), decode_first_lecture_day(chromosome), decode_first_lecture_timeslot(chromosome), decode_first_lecture_room(chromosome), room_usage, conflict_details)

    # # 2. A classroom should be big enough to accommodate the section.
    conflicts += sufficient_room_capacity(decode_section_strength(chromosome), decode_first_lecture_room_size(chromosome), decode_first_lecture_room(chromosome), conflict_details)
    conflicts += sufficient_room_capacity(decode_section_strength(chromosome), decode_second_lecture_room_size(chromosome), decode_second_lecture_room(chromosome), conflict_details)

    # # 3. A professor should not be assigned two different lectures at the same time.
    conflicts += professor_assigned_to_two_lectures_same_time(decode_first_lecture_day(chromosome), decode_first_lecture_timeslot(chromosome), population, decode_professor(chromosome), conflict_details)
    conflicts += professor_assigned_to_two_lectures_same_time(decode_second_lecture_day(chromosome), decode_second_lecture_timeslot(chromosome), population, decode_professor(chromosome), conflict_details)

    # # 4. The same section cannot be assigned to two different rooms at the same time
    conflicts += same_section_assigned_different_rooms_at_same_time(decode_first_lecture_day(chromosome), decode_first_lecture_timeslot(chromosome), decode_first_lecture_room(chromosome), population, decode_section(chromosome), conflict_details)
    conflicts += same_section_assigned_different_rooms_at_same_time(decode_second_lecture_day(chromosome), decode_second_lecture_timeslot(chromosome), decode_second_lecture_room(chromosome), population, decode_section(chromosome), conflict_details)

    # # 5. A room cannot be assigned for two different sections at the same time.
    conflicts += same_room_assigned_to_different_sections_at_same_time(decode_first_lecture_day(chromosome), decode_first_lecture_timeslot(chromosome), decode_first_lecture_room(chromosome), population, decode_section(chromosome), conflict_details)
    conflicts += same_room_assigned_to_different_sections_at_same_time(decode_second_lecture_day(chromosome), decode_second_lecture_timeslot(chromosome), decode_second_lecture_room(chromosome), population, decode_section(chromosome), conflict_details)

    # #6. No Professor can teach more than 2 courses
    conflicts += professor_course_count(decode_professor(chromosome), professor_courses, conflict_details)

    # #7. No Section can have more than 5 courses in a semester
    conflicts += section_course_count(decode_course(chromosome), decode_section(chromosome), section_courses, conflict_details)

    #8. Each course would have two lectures per week not on the same or adjacent days
    conflicts += course_have_two_lectures_per_week_not_on_same_or_adjacent_days(decode_course(chromosome), population, conflict_details) 

    # #9. Lab lectures should be conducted in two consecutive slots.
    conflicts += lab_lectures_two_consective_time_slots(decode_course_type(chromosome), decode_first_lecture_timeslot(chromosome), decode_second_lecture_timeslot(chromosome), conflict_details)

    #10. 15 mins breaks allowed between consecutive classes to ensure that there is sufficient time for transitions between classes.
    # timeslots are defined in such a way that timeslots have 15 min gaps in between

    return conflicts
    
```

```python
def fitness_function(population):
    room_usage = {}
    professor_courses = {}
    section_courses = {}
    conflict_details = []
    fitness_scores = []

    conflicts = 0
    for chromosome in population:
       conflicts = calculate_conflicts(chromosome, population, room_usage, professor_courses, section_courses, conflict_details)
       fitness_score = -conflicts
       fitness_scores.append(fitness_score)

    return fitness_scores
```

```python
from random import randint
def tournament_selection(population, fitness_scores, k = 20):

    selection_index = randint(0, len(population) - 1)
    for _ in range(k - 1):
        ix = randint(0, len(population) - 1)
        if fitness_scores[ix] > fitness_scores[selection_index]:
            selection_index = ix

    return population[selection_index]
```

```python
import random
def crossover(parent1, parent2, r_cross):
    child1, child2 = parent1.copy(), parent2.copy()
    gene_sizes = [len(Courses), 2, len(Sections), 7, len(Professors), 5, len(Timeslots), len(Rooms), 2, 5, len(Timeslots), len(Rooms)]
    num_genes = len(gene_sizes)

    if random.random() < r_cross:
        chromosone_length = len(parent1)

        gene_end_indices = [sum(gene_sizes[:i]) for i in range (1, num_genes + 1)]
        for gene_end_idx in gene_end_indices:
            crossover_point = random.randint(gene_end_idx, chromosone_length - 1)
            child1 = parent1[:crossover_point] + parent2[crossover_point:]
            child2 = parent2[:crossover_point] + parent1[crossover_point:]

    return [child1, child2]
```

```python
import random

def mutation(chromosome, r_mut):
    mutated_chromosome = chromosome[:]
    num_genes = len(Courses) + 2 + len(Sections) + 7 + len(Professors) + 5 + len(Timeslots) + len(Rooms) + 2 + 5 + len(Timeslots) + len(Rooms)
    
    for i in range(len(mutated_chromosome)):
        if random.random() < r_mut:
            mutated_chromosome[i] = 1 - mutated_chromosome[i]

    gene_start_idx = 0
    for gene_size in [len(Courses), 2, len(Sections), 7, len(Professors), 5, len(Timeslots), len(Rooms), 2, 5, len(Timeslots), len(Rooms)]:
        gene = mutated_chromosome[gene_start_idx : gene_start_idx + gene_size]
        if sum(gene) == 0:
            idx = random.randint(gene_start_idx, gene_start_idx + gene_size - 1)
            mutated_chromosome[idx] = 1
        gene_start_idx += gene_size

    return mutated_chromosome
```

```python
def create_next_generation(selected, r_cross, r_mut):
    children = []
    num_parents = len(selected)

    if num_parents % 2 == 1:
        num_parents -= 1

    for i in range(0, num_parents, 2):
        p1, p2 = selected[i], selected[i + 1]
        for child in crossover(p1, p2, r_cross):
            mutated_child = mutation(child, r_mut)
            children.append(mutated_child)

    return children
```
